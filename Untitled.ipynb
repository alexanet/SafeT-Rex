```python
R_1 = lambda delta, L : L / tan(delta)
```

```python
R = lambda delta, a2, L : sqrt(a2**2 + R_1 ( delta, L )**2)
```

```python
R_i = lambda delta, L, w : R_1 (delta, L) - w / 2
```

```python
R_o = lambda delta, L, w : R_1 (delta, L) + w / 2
```

```python
delta_i = lambda delta, L, w : atan(L / R_i (delta, L, w))
```

```python
delta_o = lambda delta, L, w : atan(L / R_o (delta, L, w))
```

```python
R_min = lambda delta, L, w, w2, g : R_1 (delta, L ) - 1/2*w2
```

```python
R_max = lambda delta, L, w, w2, g : sqrt( (R_min (delta, L, w, w2, g ) + w)**2 + (L+g)**2 )
```

```python
R_diff = lambda delta, L, w, w2, g : R_max (delta, L, w, w2, g)- R_min (delta, L, w, w2, g)
```
